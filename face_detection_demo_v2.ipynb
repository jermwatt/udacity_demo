```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
import urllib, cStringIO
from PIL import Image
```

```python
# Choose a URL with an image you would like to detect faces in
URL = 'http://cp91279.biography.com/5-Facts-About-the-Wright-Brothers_HD_768x432-16x9.jpg'

# convert image from URl to numpy array for face detection
file = cStringIO.StringIO(urllib.urlopen(URL).read())
img = Image.open(file)
img = np.asarray(img)

# plot image
fig = plt.figure(figsize = (16,5))
ax1 = fig.add_subplot(111)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.imshow(img)
```

```python
# load in cascade classifier
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

# convert go grayscale and find faces
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)

# print img with detected faces
fig = plt.figure(figsize = (16,5))
ax1 = fig.add_subplot(111)
ax1.set_xticks([])
ax1.set_yticks([])
ax1.imshow(img)
```
